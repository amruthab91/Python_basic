**1) Explain the Linear Regression briefly in the context of the problem at hand.**

**sol:** The value of a parameter can be predicted using linear regression based on the value of another parameter. It calculates the linear connection that exists between one or more independent characteristics and a variable that is dependent. Here, it is used to predict the salary based on the experience of the person. X axis is the work experience and Y-axis is the salary of a person. The best fit line for the model is found using linear regression.

**Result for the test set 20%. Accuracy observed here is 99%**

```python
from sklearn.metrics import r2_score
import numpy
import matplotlib.pyplot as plot
import pandas
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

dataset=pandas.read_csv('salary_data.csv')
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,1].values
xTrain, xTest, yTrain, yTest = train_test_split(x,y,test_size=0.2,random_state=0)

linearRegressor=LinearRegression()
linearRegressor.fit(xTrain, yTrain)
yPrediction=linearRegressor.predict(xTest)

#Mean absolute error
meanabsoluteerror = mean_absolute_error(yTest, yPrediction)
print("The Mean Absolute Error of our Model is {}".format(round(meanabsoluteerror, 2)))

#R2 score(accuracy)
score = r2_score(yTest, yPrediction)

#regression line parameters using least square method
b1=linearRegressor.coef_[0]
b0=linearRegressor.intercept_

print('slope=',b1)
print('intercept=',b0)
print("Regression Line Equation: y = {:.2f}x + {:.2f}".format(b1, b0))

plot.scatter(xTrain,yTrain,color='red')
plot.plot(xTrain,linearRegressor.predict(xTrain),color='blue')
plot.title(f"""Salary VS Experience (Training Set for 20%) \n Accuracy= {format(round(score, 2) *100)}%""")

plot.xlabel('Years of Experience')
plot.ylabel('salary')
plot.show()

```
